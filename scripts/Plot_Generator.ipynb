### Merge results

```python
import csv
import os

def merge_csvs(file1, file2, output_file):
    headers = []
    rows = []
    
    # Read first file (macOS)
    if os.path.exists(file1):
        with open(file1, 'r', newline='') as f:
            reader = csv.reader(f)
            headers = next(reader)
            rows.extend(list(reader))
    
    # Read second file (WSL)
    if os.path.exists(file2):
        with open(file2, 'r', newline='') as f:
            reader = csv.reader(f)
            file2_headers = next(reader)
            # Ensure headers match
            if file2_headers == headers:
                rows.extend(list(reader))
            else:
                print(f"Warning: Headers in {file2} do not match {file1}. Skipping.")
    
    # Write combined file
    with open(output_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(rows)
    
    print(f"Successfully merged {len(rows)} rows into {output_file}")

if __name__ == "__main__":
    macos_file = "/Users/mptb/Documents/Studia/Data_Science/Projekt_MAG/PROJEKT_MAG/results/macos_results.csv"
    wsl_file = "/Users/mptb/Documents/Studia/Data_Science/Projekt_MAG/PROJEKT_MAG/build/results.csv"
    output_file = "/Users/mptb/Documents/Studia/Data_Science/Projekt_MAG/PROJEKT_MAG/results/combined_results.csv"
    
    merge_csvs(macos_file, wsl_file, output_file)
```
